8. What is Multiple Linear Regression?




Regression with more than one predictor: . Estimates via OLS or regularized methods.


9. What is the main difference between Simple and Multiple Linear Regression?





Simple has one predictor; multiple has two or more — so coefficients represent the effect of each predictor conditional on the others.


10. What are the key assumptions of Multiple Linear Regression?





Same as simple regression plus: no (near) perfect multicollinearity among predictors; linear relationship between predictors and outcome (or linear-in-parameters).


11. What is heteroscedasticity, and how does it affect the results of a Multiple Linear Regression model?






Heteroscedasticity = non-constant variance of errors (variance depends on predictors). Effects: OLS coefficients remain unbiased but standard errors are wrong → invalid t-tests/confidence intervals. Remedies: robust (heteroscedasticity-consistent) standard errors, transform variables, or model variance explicitly.


12. How can you improve a Multiple Linear Regression model with high multicollinearity?






Remove or combine correlated predictors.

Use dimensionality reduction (PCA).

Use regularization (Ridge, Lasso).

Center/standardize variables (helps interpretation).

Gather more data if possible.

16. What is the significance of the slope in regression analysis, and how does it affect predictions?






The slope (coefficient) quantifies the expected change in the outcome per unit change in the predictor (conditional on other variables). Larger magnitude → bigger effect on predictions.


17. How does the intercept in a regression model provide context for the relationship between variables?






It anchors the predicted value at baseline (when predictors = 0). Helps produce absolute predictions (not just relative changes).

19. How would you interpret a large standard error for a regression coefficient?





Large SE → low precision of the estimate; wide confidence interval; coefficient may be statistically non-significant. Causes: small sample size, multicollinearity, or noisy data.


20. How can heteroscedasticity be identified in residual plots, and why is it important to address it?





Identification: plot residuals vs. fitted values — look for funnel shapes (variance rising or decreasing) or patterns. Important because it invalidates standard errors and hypothesis tests if ignored.

24. How does polynomial regression differ from linear regression?





Linear regression models a straight line; polynomial regression models nonlinear relationships by adding powers of predictors (still linear in coefficients).


25. When is polynomial regression used?




When the relationship between predictor(s) and outcome is nonlinear but can be well-approximated by polynomial terms.

28. What are the limitations of polynomial regression?





Prone to overfitting, especially for high degree.

Poor behavior outside data range (bad extrapolation).

Coefficients can be unstable (multicollinearity among polynomial terms).

Interpretability decreases with degree and interaction complexity.

31. How is polynomial regression implemented in Python?







Common approach: use sklearn.preprocessing.PolynomialFeatures to create polynomial terms, then fit a linear model (e.g., LinearRegression, Ridge). Use Pipeline + GridSearchCV for degree selection and regularization. See code below.

```python
# Python examples: Simple, Multiple, Polynomial regression + selection & diagnostics
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# sklearn imports
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV, train_test_split, cross_val_score
from sklearn.metrics import mean_squared_error, r2_score

# statsmodels for Breusch-Pagan test (heteroscedasticity)
import statsmodels.api as sm
from statsmodels.stats.diagnostic import het_breuschpagan

# -------------------------
# Generate synthetic data
np.random.seed(0)
n = 200
x = np.linspace(0, 10, n)
# true relation is quadratic with noise
y = 1.5 + 0.8*x - 0.12*x**2 + np.random.normal(scale=3.0, size=n)
df = pd.DataFrame({'x': x, 'y': y})

# -------------------------
# 1) Simple linear regression (closed form and sklearn)
X = df[['x']].values
Y = df['y'].values

# closed-form OLS slope and intercept
x_bar, y_bar = X.mean(), Y.mean()
m = np.sum((X.flatten()-x_bar)*(Y-y_bar)) / np.sum((X.flatten()-x_bar)**2)
c = y_bar - m*x_bar
print("Closed-form OLS slope m =", m, "intercept c =", c)

# sklearn
lr = LinearRegression()
lr.fit(X, Y)
print("sklearn coef:", lr.coef_[0], "intercept:", lr.intercept_)
print("R2 (linear):", r2_score(Y, lr.predict(X)))

# plot
plt.figure(figsize=(6,4))
plt.scatter(X, Y, s=10, label='data')
plt.plot(X, lr.predict(X), color='orange', label='linear fit')
plt.legend()
plt.xlabel('x'); plt.ylabel('y'); plt.title('Simple Linear Regression')
plt.show()

# -------------------------
# 2) Polynomial regression (create features and fit)
degree = 2
poly = PolynomialFeatures(degree=degree, include_bias=False)
X_poly = poly.fit_transform(X)  # columns: x, x^2
lr_poly = LinearRegression()
lr_poly.fit(X_poly, Y)
print("Polynomial coefficients (degree 2):", lr_poly.coef_, "intercept:", lr_poly.intercept_)
print("R2 (poly deg=2):", r2_score(Y, lr_poly.predict(X_poly)))

# Plot polynomial fit
x_plot = np.linspace(X.min(), X.max(), 300).reshape(-1,1)
y_plot = lr_poly.predict(poly.transform(x_plot))
plt.figure(figsize=(6,4))
plt.scatter(X, Y, s=10, label='data')
plt.plot(x_plot, y_plot, color='red', label='poly fit deg=2')
plt.legend()
plt.xlabel('x'); plt.ylabel('y'); plt.title('Polynomial Regression')
plt.show()

# -------------------------
# 3) Polynomial degree selection with cross-validation (pipeline + grid search)
pipe = Pipeline([
    ('poly', PolynomialFeatures(include_bias=False)),
    ('scaler', StandardScaler()),           # helps regularized models
    ('ridge', Ridge())
])
param_grid = {'poly__degree': [1,2,3,4,5], 'ridge__alpha': [0.01, 0.1, 1.0, 10.0]}
gs = GridSearchCV(pipe, param_grid, cv=5, scoring='neg_mean_squared_error')
gs.fit(X, Y)
print("Best params:", gs.best_params_)
print("Best CV MSE:", -gs.best_score_)

# Use best estimator to predict and plot
best = gs.best_estimator_
y_pred_best = best.predict(x_plot)
plt.figure(figsize=(6,4))
plt.scatter(X, Y, s=10)
plt.plot(x_plot, y_pred_best, color='green', label='best model')
plt.legend(); plt.title('Best Polynomial (CV selected)')
plt.show()

# -------------------------
# 4) Residual diagnostics (for best polynomial)
y_hat = best.predict(X)
residuals = Y - y_hat

plt.figure(figsize=(6,4))
plt.scatter(y_hat, residuals, s=10)
plt.axhline(0, color='black', linestyle='--')
plt.xlabel('Fitted values'); plt.ylabel('Residuals'); plt.title('Residuals vs Fitted')
plt.show()

# Breusch-Pagan test (requires statsmodels)
X_sm = sm.add_constant(best.named_steps['poly'].transform(X))  # design matrix used in regression
bp_test = het_breuschpagan(residuals, X_sm)
bp_labels = ['LM stat', 'LM p-value', 'F stat', 'F p-value']
print(dict(zip(bp_labels, bp_test)))
# If p-value < 0.05 → evidence of heteroscedasticity.

# -------------------------
# 5) Multiple regression example (synthetic)
np.random.seed(1)
n = 200
x1 = np.random.normal(size=n)
x2 = np.random.normal(size=n)
y_multi = 2.0 + 1.5*x1 - 0.7*x2 + 0.5*x1*x2 + np.random.normal(scale=1.0, size=n)
df2 = pd.DataFrame({'x1': x1, 'x2': x2, 'y': y_multi})

Xm = df2[['x1','x2']]
ym = df2['y']
lrm = LinearRegression()
lrm.fit(Xm, ym)
print("Multiple regression coefficients:", lrm.coef_, "intercept:", lrm.intercept_)
print("R2 (multiple):", r2_score(ym, lrm.predict(Xm)))

# If interaction is important, add x1*x2 term:
Xm_inter = Xm.copy()
Xm_inter['x1_x2'] = Xm['x1']*Xm['x2']
lrm2 = LinearRegression().fit(Xm_inter, ym)
print("With interaction coef:", lrm2.coef_, "intercept:", lrm2.intercept_)
print("R2 with interaction:", r2_score(ym, lrm2.predict(Xm_inter)))
```
